### Making a function to remove outliers by adjusting the values of the outlier between the upper boundary and lower boundary of that column

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5*IQR)
    upper_range=Q3+(1.5*IQR)
    return lower_range,upper_range
```

### Calling above funnction to remove outlier from the coloumn Electric Range

```python
lr,ur=remove_outlier(data['Electric Range'])
data['Electric Range']=np.where(data['Electric Range']>ur,ur,data['Electric Range'])
data['Electric Range']=np.where(data['Electric Range']<lr,lr,data['Electric Range'])
```

### Ploting boxplot of column Electric range to check if any outliers present after the removal

```python
plt.figure(figsize=(10,3))
sns.boxplot(data=data,x='Electric Range')
plt.show()
```

* no outlier is present in the column Electric Range now
